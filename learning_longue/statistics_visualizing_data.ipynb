### Some basic statistical visualization libraries in Python

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Scatter plots

```python
from sklearn.datasets import load_iris
# Load the Iris dataset
iris_data = load_iris()

# Convert the data to a DataFrame
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)

# Add target column
df['target'] = iris_data.target

# Map target values to target names
target_names = iris_data.target_names
df['species'] = df['target'].map(lambda x: target_names[x])

# Create scatter plot
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', data=df)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

### Swarm plot and boc plots

```python
# Create swarm plot
sns.swarmplot(x='species', y='petal length (cm)', data=df)
plt.title('Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# Create box plot
sns.boxplot(x='species', y='petal width (cm)', data=df)
plt.title('Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
# Setting box plot condition

# Set the threshold value
threshold_value = 1.5

# Create box plot only for instances where petal width is greater than the threshold value
sns.boxplot(x='species', y='petal width (cm)', data=df[df['petal width (cm)'] > threshold_value])
```
